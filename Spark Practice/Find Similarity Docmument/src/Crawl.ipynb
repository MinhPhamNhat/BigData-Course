```python
import requests
from bs4 import BeautifulSoup
from IPython.display import clear_output
```

# Crawl Vnexpress

```python
def vnexpress_crawler(category):
    res = requests.get('https://vnexpress.net/'+category)
    doc = BeautifulSoup(res.content, 'html.parser')
    try:
        article = doc.select('a[data-medium*=Item-]')
    except:
        return
    data=[]
    for i in article:
        link = i['href']
        if link not in data:
            data.append(link)
    crawl_data = []
    success = 0
    for i in range(len(data)):
        try:
            res = requests.get(data[i])
            doc = BeautifulSoup(res.content, 'html.parser')
            title = doc.find('h1' ,{'class': 'title-detail'}).text
            content = doc.find_all('p',{'class','Normal'})
            if ''.join(j.text for j in content) != '':
                crawl_data.append({'title':title, 'content': ''.join(j.text for j in content).replace('\n','')})
                success += 1
        except:
            continue
        clear_output(wait = True)
        print("Crawl successfully {} out of {} data".format(success, len(data)))
    return crawl_data
```

# Crawl Vietnamnet

```python
def vietnamnet_crawler(category):
    res = requests.get('https://vietnamnet.vn/vn/'+category)
    doc = BeautifulSoup(res.content, 'html.parser')
    try:
        article = doc.find_all('a',{'class':'f-18 title'})
    except:
        return
    data=[]
    for i in article:
        link = i['href']
        if link not in data:
            if 'https://vietnamnet.vn/' not in link:
                data.append('https://vietnamnet.vn/'+link)
            else:
                data.append(link)
    crawl_data = []
    success = 0
    for i in range(len(data)):
        try:
            res = requests.get(data[i])
            doc = BeautifulSoup(res.content, 'html.parser')
            title = doc.find('h1' ,{'class': 'title f-22 c-3e'}).text
            content = doc.select('div[class=ArticleContent]>p')
            if ''.join(j.text for j in content) != '':
                crawl_data.append({'title':title, 'content': ''.join(j.text for j in content).replace('\n','')})
                success += 1
        except:
            continue
        clear_output(wait = True)
        print("Crawl successfully {} out of {} data".format(success, len(data)))
    return crawl_data
```

# Crawl Thanhnien

```python
def thanhnien_crawler(category):
    res = requests.get('https://thanhnien.vn/'+category)
    doc = BeautifulSoup(res.content, 'html.parser')
    try:
        article = doc.find_all('a',{'class':'story__title'})
    except:
        return
    data=[]
    for i in article:
        link = i['href']
        if link not in data:
            if 'https://thanhnien.vn/' not in link:
                data.append('https://thanhnien.vn/'+link)
            else:
                data.append(link)
    crawl_data = []
    success = 0
    for i in range(len(data)):
        try:
            res = requests.get(data[i])
            doc = BeautifulSoup(res.content, 'html.parser')
            title = doc.find('h1' ,{'class': 'details__headline'}).text
            content = doc.select('div[itemprop=articleBody]>div')
            if ''.join(j.text for j in content) != '':
                crawl_data.append({'title':title, 'content': ''.join(j.text for j in content).replace('\n','')})
                success += 1
        except:
            continue
        clear_output(wait = True)
        print("Crawl successfully {} out of {} data".format(success, len(data)))
    return crawl_data
```

# Crawl tuoitre

```python
def tuoitre_crawler(category):
    res = requests.get('https://tuoitre.vn/'+category+'.htm')
    doc = BeautifulSoup(res.content, 'html.parser')
    try:
        article = doc.find_all('a',{'data-linktype':'newsdetail'})
    except:
        return
    data=[]
    for i in article:
        link = i['href']
        if link not in data:
            if 'https://tuoitre.vn/' not in link:
                data.append('https://tuoitre.vn/'+link)
            else:
                data.append(link)
    crawl_data = []
    success = 0
    for i in range(len(data)):
        try:
            res = requests.get(data[i])
            doc = BeautifulSoup(res.content, 'html.parser')
            title = doc.find('h1' ,{'class': 'article-title'}).text
            content = doc.select('div[id=main-detail-body]>p')
            if ''.join(j.text for j in content) != '':
                crawl_data.append({'title':title, 'content': ''.join(j.text for j in content).replace('\n','')})
                success += 1
        except:
            continue
        clear_output(wait = True)
        print("Crawl successfully {} out of {} data".format(i+1, len(data)))
    return crawl_data
```
